```python
import pandas as pd
target_df = pd.read_csv("target.csv",index_col=0)
print(target_df.shape)
```

```python
def reset_table(df):
    df = df.drop(range(0, 3), axis=0)
    cname = df[0:1].values[0] 
    df = df.reset_index()
    df = df.drop(range(0, 1), axis=0)
    df = df.drop('index', axis=1)
    df.columns = cname
    return df

def oil_preprocess(df, start_year = 1982):
    years = []
    for i, v in df.iterrows():
        if pd.isnull(v[1]):
            years.append(year)
        else:
            year = int(str(v[1]).strip('年'))
            if len(str(year))!=4:
                year = int(str(year)[:4])
            years.append(year)
            start_year += 1
    df.insert(1, 'year', years)  
    df = df.drop(df.columns[2], axis=1)
    return df.fillna(0)

df = reset_table(target_df)
df = oil_preprocess(df, start_year = 1982)
df.to_csv("target_after.csv", header=True, index=True)
```
